**Q1: Find the null values and fill it with the mean of that column (Data Cleaning)**

```python
cars_data.isnull().sum() #shows the null values in each column
```

```python
# Fill null values in numeric columns with the mean of each column
cars_data.fillna(cars_data.select_dtypes(include=['number']).mean(), inplace=True)
```

```python
cars_data.fillna(cars_data.select_dtypes(exclude=["number"]).apply(lambda x: "Unknown"), inplace = True) #To fill non numeric columns with the term Unknown
```
